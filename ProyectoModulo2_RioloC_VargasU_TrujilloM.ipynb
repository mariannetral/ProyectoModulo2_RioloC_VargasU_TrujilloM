### 2.4 Solución del problema de optimización.



```python
import yfinance as yf
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getClose(name,start,end):   #para meter varios tickers
    return yf.download(name,start=start,end=end)["Adj Close"]
```


**DATOS PRE PANDEMIA**


```python
dataVTI1 = getClose(["VTI"], "2018-01-01","2019-12-31")
dataSPY1 = getClose(["SPY"],"2018-01-01","2019-12-31")
dataVYM1 = getClose(["VYM"], "2018-01-01","2019-12-31")
dataVBK1 = getClose(["VBK"],"2018-01-01","2019-12-31")
dataXLY1 = getClose(["XLY"],"2018-01-01","2019-12-31")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

plt.figure()
dataVTI1.plot(label = "VTI1")
dataSPY1.plot(label = "SPY1")
dataVYM1.plot(label = "VYM1")
dataVBK1.plot(label = "VBK1")
dataXLY1.plot(label = "XLY1")

plt.legend()
plt.grid()
```

**DATOS POST PANDEMIA**

```python
dataVTI2= getClose(["VTI"], "2020-01-01","2021-12-31")
dataSPY2 = getClose(["SPY"],"2020-01-01","2021-12-31")
dataVYM2= getClose(["VYM"], "2020-01-01","2021-12-31")
dataVBK2 = getClose(["VBK"],"2020-01-01","2021-12-31")
dataXLY2 = getClose(["XLY"],"2020-01-01","2021-12-31")
```

```python
plt.figure()
dataVTI2.plot(label = "VTI2")
dataSPY2.plot(label = "SPY2")
dataVYM2.plot(label = "VYM2")
dataVBK2.plot(label = "VBK2")
dataXLY2.plot(label = "XLY2")


plt.legend()
plt.grid()
```

**CALCULAMOS LOS RENDIMIENTOS SIMPLES DEL PERIODO PRE PANDEMIA DE CADA ETF**

```python
R_VTI1 = (dataVTI1 - dataVTI1.shift())/dataVTI1.shift()
R_SPY1 = (dataSPY1 - dataSPY1.shift())/dataSPY1.shift()
R_VYM1 = (dataVYM1 - dataVYM1.shift())/dataVYM1.shift()
R_VBK1 = (dataVBK1 - dataVBK1.shift())/dataVBK1.shift()
R_XLY1 = (dataXLY1 - dataXLY1.shift())/dataXLY1.shift()
```

**SIMULAMOS 1000 ESCENARIOS DE RENDIMIENTOS DIARIOS PARA EL PERIODO 2018-2019 PARA CADA ETF**

```python
RVTI1_mean = R_VTI1.mean()
RVTI1_std = R_VTI1.std()

RSPY1_mean = R_SPY1.mean()
RSPY1_std = R_SPY1.std()

RVYM1_mean = R_VYM1.mean()
RVYM1_std = R_VYM1.std()

RVBK1_mean = R_VBK1.mean()
RVBK1_std = R_VBK1.std()

RXLY1_mean = R_XLY1.mean()
RXLY1_std = R_XLY1.std()
```

```python
def sim_r(media,std,inicio,final,N):
   
    
    r= np.random.normal(media,std,size=(len(pd.date_range(inicio, final)), N))
    fechas= pd.date_range(inicio,final)
    return pd.DataFrame(index=fechas, data=r)
```

```python
SIMULADOS_VTI1 = sim_r(RVTI1_mean,RVTI1_std,"2018-01-01","2019-12-31",1000)
SIMULADOS_SPY1 = sim_r(RSPY1_mean,RSPY1_std,"2018-01-01","2019-12-31",1000)
SIMULADOS_VYM1 = sim_r(RVYM1_mean,RVYM1_std,"2018-01-01","2019-12-31",1000)
SIMULADOS_VBK1 = sim_r(RVBK1_mean,RVBK1_std,"2018-01-01","2019-12-31",1000)
SIMULADOS_XLY1 = sim_r(RXLY1_mean,RXLY1_std,"2018-01-01","2019-12-31",1000)

```

**Calcule los precios con base en los rendimientos simulados**

```python
preciosVTI1_simulados=(SIMULADOS_VTI1 +1).cumprod()*dataVTI1[-1]
preciosSPY1_simulados=(SIMULADOS_SPY1 +1).cumprod()*dataSPY1[-1]
preciosVYM1_simulados=(SIMULADOS_VYM1 +1).cumprod()*dataVYM1[-1]
preciosVBK1_simulados=(SIMULADOS_VBK1 +1).cumprod()*dataVBK1[-1]
preciosXLY1_simulados=(SIMULADOS_XLY1 +1).cumprod()*dataXLY1[-1]


```

 **Gráfica del promedio de los precios simulados de ambas acciones**

```python
plt.figure()


preciosVTI1_simulados.mean(axis=1).plot(label = "VTI SIMULADO")
preciosSPY1_simulados.mean(axis=1).plot(label = "SPY SIMULADO")
preciosVYM1_simulados.mean(axis=1).plot(label = "VYM SIMULADO")
preciosVBK1_simulados.mean(axis=1).plot(label = "VBK SIMULADO")
preciosXLY1_simulados.mean(axis=1).plot(label = "XLY SIMULADO")

plt.legend()

plt.grid()
```

Con base en nuestras simulaciones notamos que SPY es el EFT que tiene un mayor crecimiento en el precio, por lo que optaremos por tomar a SPY como nuestra mejor opción para invertir.


### 2.5 Visualización de la solución del problema.






**Finalmente, grafique los datos reales del periodo PRE PANDEMIA para ver que tan acertado fue nuestro análisis.**

```python

plt.figure()
dataVTI1.plot(label = "VTI", color="green")
dataSPY1.plot(label = "SPY", color="blue")
dataVYM1.plot(label = "VYM", color="black")
dataVBK1.plot(label = "VBK", color="cyan")
dataXLY1.plot(label = "XLY", color="orange")

preciosVTI1_simulados.mean(axis=1).plot(label = "VTI SIMULADO", color="green")
preciosSPY1_simulados.mean(axis=1).plot(label = "SPY SIMULADO", color="blue")
preciosVYM1_simulados.mean(axis=1).plot(label = "VYM SIMULADO", color="black")
preciosVBK1_simulados.mean(axis=1).plot(label = "VBK SIMULADO", color="cyan")
preciosXLY1_simulados.mean(axis=1).plot(label = "XLY SIMULADO", color="orange")



plt.legend()
plt.grid()
plt.show()
```

### ¿En qué ETF invertir de acuerdo a los datos de las gráficas?

```python
VTI_rend_ini=np.array(preciosVTI1_simulados)[1]
VTI_rend_fin=np.array(preciosVTI1_simulados)[-1]
VTI_rend=np.array((VTI_rend_fin.mean())/(VTI_rend_ini.mean()))

SPY_rend_ini=np.array(preciosSPY1_simulados)[1]
SPY_rend_fin=np.array(preciosSPY1_simulados)[-1]
SPY_rend=np.array((SPY_rend_fin.mean())/(SPY_rend_ini.mean()))

VYM_rend_ini=np.array(preciosVYM1_simulados)[1]
VYM_rend_fin=np.array(preciosVYM1_simulados)[-1]
VYM_rend=np.array((VYM_rend_fin.mean())/(VYM_rend_ini.mean()))

VBK_rend_ini=np.array(preciosVBK1_simulados)[1]
VBK_rend_fin=np.array(preciosVBK1_simulados)[-1]
VBK_rend=np.array((VBK_rend_fin.mean())/(VBK_rend_ini.mean()))

XLY_rend_ini=np.array(preciosXLY1_simulados)[1]
XLY_rend_fin=np.array(preciosXLY1_simulados)[-1]
XLY_rend=np.array((XLY_rend_fin.mean())/(XLY_rend_ini.mean()))

rend1=(VTI_rend, SPY_rend, VYM_rend, VBK_rend, XLY_rend)

plt.figure()
y=[1,2,3,4,5]
plt.scatter(rend1, y)
```

La gráfica con indica que el ETF "XLY" es el que tuvo el mayor rendimiento, con un incremento del $ 1.497x $ respecto a su precio a inicios de 2018

### 2.4 Solución del problema de optimización. (pt2)


Ahora Calculamos los Rendimientos Simples para el periodo Post-Pandemia


```python
R_VTI2 = (dataVTI2 - dataVTI2.shift())/dataVTI2.shift()
R_SPY2 = (dataSPY2 - dataSPY2.shift())/dataSPY2.shift()
R_VYM2 = (dataVYM2 - dataVYM2.shift())/dataVYM2.shift()
R_VBK2 = (dataVBK2 - dataVBK2.shift())/dataVBK2.shift()
R_XLY2 = (dataXLY2 - dataXLY2.shift())/dataXLY2.shift()
```

Extraemos la Media y la Desviación Estandar para poder Simular Rendimientos a continuación

```python
RVTI2_mean = R_VTI2.mean()
RVTI2_std = R_VTI2.std()

RSPY2_mean = R_SPY2.mean()
RSPY2_std = R_SPY2.std()

RVYM2_mean = R_VYM2.mean()
RVYM2_std = R_VYM2.std()

RVBK2_mean = R_VBK2.mean()
RVBK2_std = R_VBK2.std()

RXLY2_mean = R_XLY2.mean()
RXLY2_std = R_XLY2.std()
```

Simulamos 1000 escenarios entre el 01/01/2020 y 31/12/2021


```python
SIM_VTI2 = sim_r(RVTI2_mean,RVTI2_std,"2020-01-01","2021-12-31",1000)
SIM_SPY2 = sim_r(RSPY2_mean,RSPY2_std,"2020-01-01","2021-12-31",1000)
SIM_VYM2 = sim_r(RVYM2_mean,RVYM2_std,"2020-01-01","2021-12-31",1000)
SIM_VBK2 = sim_r(RVBK2_mean,RVBK2_std,"2020-01-01","2021-12-31",1000)
SIM_XLY2 = sim_r(RXLY2_mean,RXLY2_std,"2020-01-01","2021-12-31",1000)
```

Ahora Calculamos los precios de los escenarios simulados

```python
preciosVTI2_sim=(SIM_VTI2 +1).cumprod()*dataVTI2[-1]
preciosSPY2_sim=(SIM_SPY2 +1).cumprod()*dataSPY2[-1]
preciosVYM2_sim=(SIM_VYM2 +1).cumprod()*dataVYM2[-1]
preciosVBK2_sim=(SIM_VBK2 +1).cumprod()*dataVBK2[-1]
preciosXLY2_sim=(SIM_XLY2 +1).cumprod()*dataXLY2[-1]
```

Graficamos el Promedio de los Precios Simulados 

```python
plt.figure()

dataVTI2.plot(label = "VTI", color="green")
dataSPY2.plot(label = "SPY", color="blue")
dataVYM2.plot(label = "VYM", color="black")
dataVBK2.plot(label = "VBK", color="cyan")
dataXLY2.plot(label = "XLY", color="orange")

preciosVTI2_sim.mean(axis=1).plot(label = "VTI SIMULADO 2", color="green")
preciosSPY2_sim.mean(axis=1).plot(label = "SPY SIMULADO 2", color="blue")
preciosVYM2_sim.mean(axis=1).plot(label = "VYM SIMULADO 2", color="black")
preciosVBK2_sim.mean(axis=1).plot(label = "VBK SIMULADO 2", color="cyan")
preciosXLY2_sim.mean(axis=1).plot(label = "XLY SIMULADO 2", color="orange")


plt.legend()
plt.grid()
plt.show()
```
